```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
%matplotlib inline
```

# Relationships between bank client attributes

```python
bank_marketing = pd.read_csv('bank-full.csv', sep=';')
```

```python
bank_marketing.head(10)
```

## Plot client attributes against each other

```python
# Plot age against all continuous attributes. Not sorted in order - this isn't a histogram. However, maybe sorting can give weird predictive power.

# age vs. balance
plt.figure(figsize=(10, 5))
plt.scatter(bank_marketing.age, bank_marketing.balance)
plt.xlabel('Ages', fontsize=20)
plt.ylabel('Balance', fontsize=20)
plt.show()
print('corr coef: ', np.corrcoef(bank_marketing.age, bank_marketing.balance)[0,1])

# age vs. day
plt.figure(figsize=(10, 5))
plt.scatter(bank_marketing.age, bank_marketing.day)
plt.xlabel('Ages', fontsize=20)
plt.ylabel('Day', fontsize=20)
plt.show()
print('corr coef: ', np.corrcoef(bank_marketing.age, bank_marketing.day)[0,1])

# age vs. duration
plt.figure(figsize=(10, 5))
plt.scatter(bank_marketing.age, bank_marketing.duration)
plt.xlabel('Ages', fontsize=20)
plt.ylabel('Duration', fontsize=20)
plt.show()
print('corr coef: ', np.corrcoef(bank_marketing.age, bank_marketing.duration)[0,1])
```
